# 2. What are the top 3 movies with highest nett profit in 2013?

```python
net = imdb[['movie_title', 'title_year', 'gross_USD', 'net_USD', 'profitable']]
net = net.loc[imdb["title_year"] == 2013].sort_values(by='net_USD', ascending=False)
net.head(3)
```

# 3. What are the top 3 movies with highest gross profit in 2009?

```python
profit = imdb[imdb.title_year == 2009].sort_values(by='gross_USD', ascending=False)

profit.head(3)
```

# 6. What are the top 3 countries with the highest nett profit in all dataset times?

```python
country = imdb.groupby('country').sum().sort_values('net_USD', ascending=False)
country.head(3)
```

# 7. What are the average budget of all movies in dataset in 2013?

```python
mean = imdb[imdb.title_year == 2013]
mean.budget_USD.mean()
```

# 8. What are the average budget of all movies in dataset in 2014?

```python
mean = imdb[imdb.title_year == 2014]
mean.budget_USD.mean()
```

# 9. Which one is higher, average budget of all movies in dataset in 2014 or 2013

```python
if (imdb[imdb.title_year == 2013]['budget_USD'].mean() > imdb[imdb.title_year == 2014]['budget_USD'].mean()):
    print('2013 budgets is higher than 2014 budget')
else:
    print('2014 budgets is higher than 2013 budget')
```
